```python
from ydata_profiling import ProfileReport
import pandas as pd
```

```python
raw = pd.read_csv(
    "20231201-103525_nfi_dataset_raw copy.csv",
    index_col=[0],
    low_memory=False,
)
```

```python
all_tree_lvl_vars = [
    # From NFI data
    "acci",
    "cible",
    "clon",
    "decoupe",
    "deggib",
    "espar",
    "forme",
    "ori",
    "qualite",
    "simplif",
    "tetard",
    "tige",
    "veget",
    "veget5",
    "a",
    "datemort",
    "lib",
    "age",
    "age13",
    "c0",
    "c13",
    "ddec",
    "hdec",
    "hrb",
    "htot",
    "ir1",
    "ir5",
    "lfsd",
    "mes_c13",
    "q1",
    "q2",
    "q3",
    "r",
    "v",
    "w",
    "mortb",
    "sfcoeur",
    "sfpied",
    "sfdorge",
    "sfgeliv",
    "sfgui",
    "espar_bm",
    "frepli",
    "dbm",
    "decomp",
    "espar_c",
    "p1525",
    "p7ares",
    "strate",
    "tca",
    "tcl",
    # From derivatives
    "alive_but_injured_1",
    "alive_but_injured_2",
    "family_lat",
    "genus_lat",
    "mode_of_death",
    "order_lat",
    "revisit_state",
    "shadow_growth",
    "species_lat",
    "tree_class",
    "tree_id",
    "tree_state_1",
    "tree_state_2",
    "tree_state_change",
    "espar_red",
    "avg_growth_ba_per_year",
    "avg_growth_height_meter_per_yr",
    "ba_1",
    "ba_2",
    "ba_change_abs_yr",
    "ba_change_perc_yr",
    "c13_change_abs_yr",
    "c13_change_perc_yr",
    "dbh_1",
    "dbh_2",
    "dbh_change_abs_yr",
    "dbh_change_perc_yr",
    "visit_1",
    "visit_2",
    "age_class",
    "height_class",
    "census_interval",
]

# Attach suffixes to column names
suffix_1 = [suffix + "_1" for suffix in all_tree_lvl_vars]
suffix_2 = [suffix + "_2" for suffix in all_tree_lvl_vars]

# Create a list of all tree level variables
all_tree_lvl_vars = all_tree_lvl_vars + suffix_1 + suffix_2
```

```python
raw_tree = raw.copy()

for var in raw_tree.columns:
    if var in ["idp", "campagne_1"]:
        next
    if var not in all_tree_lvl_vars:
        raw_tree = raw_tree.drop(columns=[var], errors="ignore")
```

```python
raw_loc = raw.copy()

for var in raw_loc.columns:
    if var in ["idp", "campagne_1"]:
        next
    if var in all_tree_lvl_vars:
        raw_loc = raw_loc.drop(columns=[var], errors="ignore")

raw_loc = raw_loc.drop_duplicates()
```

```python
print(f"All raw data: {raw.shape}")
print(f"Tree-level data: {raw_tree.shape}")
print(f"Location-level data: {raw_loc.shape}")
print(f"Number of unique locations: {raw['idp'].nunique()}")
```

```python
ProfileReport(raw_tree, title="Tree-level data", minimal=True, dark_mode=True).to_file(
    "report_tree_level_raw_data.html"
)
ProfileReport(
    raw_loc, title="Location-level data", minimal=True, dark_mode=True
).to_file("report_location_level_raw_data.html")
```
